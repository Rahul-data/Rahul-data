```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://www.mea.gov.in/vaccine-supply.htm"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
# Get the title
title = soup.title
print(title)
```

```python
# Print out the text
text = soup.get_text()
print(soup.text)
```

```python
soup.find_all('table')
```

```python
rows = soup.find_all('tr')
print(rows[:])
```

```python
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
import re
import pandas as pd
list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
df = pd.DataFrame(list_rows)
df.head(10)
```
